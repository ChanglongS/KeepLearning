```python
x = np.arange(1, 100)
fig, axes = plt.subplots(2, 2)
ax1 = axes[0, 0]
ax2 = axes[1, 0]
ax3 = axes[1, 1]
ax4 = axes[0, 1]

fig = plt.figure(figsize=(20, 8), dpi=80)
ax1.plot(x, -x)
ax2.plot(x, x ** 2)
ax3.plot(x, 1 / x)
ax4.plot(x, np.log(x))
plt.show()
```
